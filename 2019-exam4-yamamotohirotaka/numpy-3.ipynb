xが-2πから2πの時のsin(x)を描画して下さい。

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np  # 前の課題と同様にキーワードを入れてください。
```

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
```

```python
np.sin(x)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, np.sin(x))  # サインカーブを出力するように□の中に適切なメソッドを入れてください。
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.plot(x, np.sin(x))  # □には、上記で出力したときと同じメソッドが入ります。
ax2.plot(x, 2*np.sin(x))  # 1つ目の□には、上記で出力したときと同じメソッドが入ります。2つ目の□には振幅を2倍にするための記述があります。数値と演算子の組み合わせを入れてください。
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, np.sin(x))  # 上記同様に□を埋めてください。
ax.plot(x, 2*np.sin(x))  # 上記同様に□を埋めてください。
plt.show()
```
